```python
#3. Data Preprocessing
#Handling Missing Values
df.fillna(df.mean(), inplace=True)
```

```python
# Check for anomalies (e.g., values that are too high or too low)
anomalies = df[(df['Ex_rate'] < df['Ex_rate'].quantile(0.05)) | (df['Ex_rate'] > df['Ex_rate'].quantile(0.95))]
print(anomalies)
```

```python
# This is a simple example; you may want to use more sophisticated methods
df.drop(anomalies.index, inplace=True)
```

```python
df=df[(df['Ex_rate']>0) & (df['Ex_rate']<2)]
df['Ex_rate'].mean()
```

```python
#1. Parameter Selection for ARIMA
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
```

```python
#Use ACF and PACF plots to determine the parameters.
# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plot_acf(df['Ex_rate'], lags=40, title='Autocorrelation Function (ACF)')
plot_pacf(df['Ex_rate'], lags=40, title='Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plot_acf(df['date'], lags=40, title='Autocorrelation Function (ACF)')
plot_pacf(df['date'], lags=40, title='Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
#2. Model Fitting
#Fit the ARIMA model using the selected parameters (p, d, q).
model = ARIMA(df['Ex_rate'], order=(1, 1, 1))
model_fit = model.fit()
model_fit.summary()
model_fit.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
model_fit.summary()
```

```python
#4. Forecasting
#Perform out-of-sample forecasting.
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)
forecast
```

```python
#plot
forecast=model_fit.forecast(steps=forecast_steps)
plt.figure(figsize=(12, 6))
plt.plot(df['Ex_rate'], label='Actual')
plt.plot(forecast, label='Forecast', color='red', linestyle='--', marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black', markeredgewidth=1, alpha=0.7)
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()
```

```python
#Part 3: Model Building - Exponential Smoothing
#model selection 1
#Choose an appropriate Exponential Smoothing model.
## Fit Holt-Winters model (example)
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model = ExponentialSmoothing(df['Ex_rate'], trend='add', seasonal='add', seasonal_periods=12)
model_fit = model.fit()
model_fit.summary()
```

```python
#2. Parameter Optimization
#Use AIC for parameter optimization.
print(f'AIC: {model_fit.aic}')
print(f'BIC: {model_fit.bic}')
print(f'SSE: {model_fit.sse}')
```

```python
#3. Model Fitting and Forecasting
#Fit the model and visualize the forecasts.
# Forecasting with Holt-Winters
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)
plt.figure(figsize=(12, 6))
plt.plot(df['Ex_rate'], label='Actual')
plt.plot(forecast, label='Forecast', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Holt-Winters Forecast')
plt.legend()
plt.show()
```

```python
#Part 4: Evaluation and Comparison
#Calculate MAE, RMSE, and MAPE.
from sklearn.metrics import mean_absolute_error, mean_squared_error
actual_values = df['Ex_rate'].tail(forecast_steps)  # Get the last 'forecast_steps' values
print("Actual Values:")
print(actual_values)  # Print the actual values
```

```python
mae = mean_absolute_error(actual_values, forecast)
rmse = mean_squared_error(actual_values, forecast, squared=False)
mape = np.mean(np.abs((actual_values - forecast) / actual_values)) * 100
print("Forecast Values:")
print(forecast)  # Print the forecast values
print("Actual Values:")
print(actual_values)  # Print the actual values
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}')
print(f'SSE: {model_fit.sse}')
```
